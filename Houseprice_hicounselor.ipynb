```python
# --- WRITE YOUR CODE FOR MODULE 3 TASK 2 ---
counts = {}
index = 0
fig, axs = plt.subplots(2,4, figsize=(12, 12))
fig.suptitle('Categorical Features: Distribution')
for i,col in enumerate(cat_cols, start=1):
  counts[col] = pd.value_counts(data[col])
  ax = axs[index // 4, index % 4]
  ax.bar(counts[col].index, counts[col].values)
  ax.set_title(col)
  ax.tick_params(axis='x', labelrotation = 90)
  ax.set_xticklabels(counts[col].index, rotation=90)
  # ax.xaxis.set_major_locator(dates.DayLocator())
  ax.bar_label(ax.containers[0]);
  index += 1
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout
plt.show();


for col in cat_cols:
    value_counts = pd.value_counts(data[col])
    col_dict = {category: count for category, count in value_counts.items()}  # Create inner dictionary
    counts[col] = col_dict  # Add inner dictionary to outer dictionary

# 'counts' 
print(counts)
```

```python
# --- WRITE YOUR CODE FOR MODULE 4 TASK 1 ---

df_encoded= pd.get_dummies(data, columns = cat_cols)

#--- Inspect data ---
from sklearn import set_config
set_config(transform_output='pandas')
pd.set_option('display.max_colwidth', 250)
df_encoded
```

```python
df_encoded.columns
```

```python
#--- Import train_test_split from sklearn.model_selection ---
from sklearn.model_selection import train_test_split

# --- WRITE YOUR CODE FOR MODULE 4 TASK 2 ---

X = df_encoded.drop('SalePrice',axis = 1)
Y = df_encoded['SalePrice']
X_train, X_valid, Y_train, Y_valid = train_test_split(X,Y, test_size = 0.2, random_state = 0)

#--- Inspect data ---
X_train.shape, X_valid.shape
Y_train.shape,Y_valid.shape
```

```python
#--- Import LinearRegression,RandomForestRegressor,svm,Ridge, mean_absolute_percentage_error,from sklearn ---
from sklearn.linear_model import LinearRegression ,Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_percentage_error

#Intilize and fit Linear Regression model and find mape for lr model
lr = LinearRegression()
lr.fit(X_train, Y_train)

lr_pred = lr.predict(X_valid)
lr_mape = mean_absolute_percentage_error(Y_valid, lr_pred) * 100

#Intilize  and fit Random Forest model and find mape for rf model
rf = RandomForestRegressor(n_estimators=10)
rf.fit(X_train, Y_train)

rf_pred = rf.predict(X_valid)
rf_mape = mean_absolute_percentage_error(Y_valid, rf_pred) 


#Intilize  and fit svm model and find mape for svr model
svr = SVR()
svr.fit(X_train, Y_train)
svr_pred = svr.predict(X_valid)
svr_mape = mean_absolute_percentage_error(Y_valid, svr_pred)


#Intilize  and fit Ridge Regression model and find mape for ridge model
rdg = Ridge()
rdg.fit(X_train, Y_train)
rdg_pred = rdg.predict(X_valid)
ridge_mape = mean_absolute_percentage_error(Y_valid, rdg_pred)
```
